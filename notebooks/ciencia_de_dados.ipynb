# **Análise de Dados via Pandas**
---

Uma empresa contratou uma equipe (que são vocês) para analisar uma planilha de cancelamentos de planos dessa empresa. O objetivo da equipe é analisar a planilha para verificar o porquê dos cancelamentos, e o que pode ser feito à respeito para diminuir o número de cancelamentos.

A primeira coisa a ser feita é ler a planilha usando o pandas.

```python
import pandas as pd

tabela = pd.read_csv("datasets/cancelamentos.csv")
display(tabela)
```

Vamos retirar a coluna ***CustomerID*** da consulta, pois ela não é interessante para a nossa análise.

```python
# elimina a coluna CustomerID da consulta
tabela = tabela.drop("CustomerID", axis=1)
display(tabela)
```

Vamos analisar se esssa tabela possui algum valor nulo, e para isso, vamos verificar o número de linhas da tabela.

```python
display(tabela.info())
```

Repare que o número total de entradas não bate com o número de valores não nulos de cada uma das colunas, e nem o número de linhas com valores não-nulos é o mesmo para cada uma das colunas.

Precisaremos eliminar os valores nulos da tabela.

```python
# elimina os valores nulos
tabela = tabela.dropna()

display(tabela.info())
```

## Análise de dados
---

Agora que fizemos o tratamento de dados, a análise de dados propriamente dita vai começar. Primeiramente, vamos verificar a quantidade de clientes que cancelaram o contrato versus a quantidade de clientes que ainda estão com o contrato vigente.

```python
# retorna o total de cancelamentos e o total de contratos ainda vigentes
display(tabela['cancelou'].value_counts())
```

A linha com o valor 1.0 representa os cancelamentos, enquanto a linha com o valor 0.0 representa os contratos ainda vigentes.

Podemos perceber que o número de cancelamentos é maior do que o número de contratos ainda vigentes, oque significa que a empresa perdeu a maioria de seus clientes. A pergunta agora é: por quê?

Antes de respondermos a essa pergunta, vamos verificar a porcentagem de cada um dos valores.

```python
display(tabela['cancelou'].value_counts(normalize=True).map("{:.2%}".format))
```

## Montando o perfil de cancelamento da empresa
---

Vamos tentar descobrir qual o perfil de cancelamento da empresa, e para isso, vamos analisar os tipos de contratos existentes na empresa com base no período de cada contrato.

```python
display(tabela['duracao_contrato'].value_counts())
display(tabela['duracao_contrato'].value_counts(normalize=True).map("{:.2%}".format))
```

Vamos agrupar as informações da coluna **duracao_contrato** e fazer a média das informações.

```python
display(tabela.groupby('duracao_contrato').mean(numeric_only=True))
```

Repare que **exatos 100%** dos clientes que optaram pelo contrato mensal cancelaram o contrato. Logo, sabemos agora que esse tipo de contrato é problemático, e portanto, precisaremos eliminá-lo e exibir um relatório sem os contratos mensais.

```python
tabela = tabela[tabela["duracao_contrato"] != "Monthly"]

display(tabela)
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format))
```

A proporção de cancelamentos agora caiu para 46,05%. Porém, esse número ainda é muito alto. Então precisaremos ir mais a fundo e ver onde podemos eliminar mais para diminuir mais ainda essa taxa.

## Gráficos
---

Uma das melhores formas de se analisar dados é através de gráficos, pois fica muito mais fácil visualizar os dados o obter informações.

Para conseguirmos fazer isso, precisaremos instalar uma biblioteca gráfica para analisar cada uma das colunas. Vamos usar a biblioteca plotly, instalando-a através do comando `pip install plotly` no terminal.

```python
import plotly.express as px

for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelou", width=600)
    grafico.show()
```

Podemos verificar que as taxas de cancelamentos são de 100% para os clientes que:

1. Precisaram ligar para o call center por mais de 5 vezes.
2. Atrasaram o pagamento da assinatura por mais de 20 dias.

Logo, vamos eliminar da consulta todos os clientes que ligaram mais de 5 vezes para o call center, e também todos os que atrasaram o pagamento por 20 dias ou mais.

```python
tabela = tabela[tabela["ligacoes_callcenter"] < 5]
tabela = tabela[tabela["dias_atraso"] <= 20]

display(tabela)
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format))
```
